```python
import scipy as sc
from scipy.stats import norm 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import linear_model
import statsmodels.formula.api as sm
%matplotlib inline
#plt.rcParams['figure.figsize']=(10,6)
advert=pd.read_csv("C:\\Users\\arora\\Desktop\\Tools and Techniques\\Python\\Advertising.csv")
advert
plt.scatter(x=advert['TV'], y=advert['sales'], c='r', marker='s', label='TV')
plt.scatter(x=advert['radio'], y=advert['sales'], c='b', marker='o', label='radio')
plt.scatter(x=advert['newspaper'], y=advert['sales'], c='k', marker='*', label='newspaper')
plt.show()
advert.describe()
regress= sm.ols(formula="sales~ TV", data=advert).fit()
regress.summary()
plt.scatter(x=advert['TV'], y=advert['sales'], c='k', marker="*", label='TV')
plt.plot(advert['TV'], regress.predict(), color='blue', linewidth=3)
plt.xlabel("TV Budget in 1000 dollars")
plt.ylabel("Sales")
plt.legend(numpoints=2, loc=4)
plt.show()
regress.ssr
regressallvar=sm.ols(formula="sales~ TV+radio+newspaper", data=advert).fit()
regressallvar.summary()
print(advert.corr())
pd.plotting.scatter_matrix(advert, figsize=(10,10), diagonal= 'kde')
plt.show()
```
