```python
import pandas as pd

import scipy as sc
from scipy.stats import bernoulli
from scipy.stats import binom
from scipy.stats import norm
from scipy.stats import t

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
plt.style.use('ggplot')
n =10
p = 0.5
f = lambda k: binom.pmf(k, n=n,p=p)

x = sc.arange(n+1);
plt.plot(x, f(x),'bo')
plt.title("The probability mass function for a Binomial random variable")
plt.xlim([0,n])
plt.ylim([0,0.5])
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.show()
```

```python
import pandas as pd

import scipy as sc
from scipy.stats import norm


import matplotlib.pyplot as plt
mu=0
sigma=1
x=sc.arange(mu-4*sigma,mu+4*sigma,0.001)
pdf=norm.pdf(x,loc=mu , scale=sigma)
plt.title("Normal / Gaussian Distribution")
plt.ylim(0,0.6)
plt.ylabel("Normal variable trials")
plt.plot(x,pdf,linewidth=2,color='b')
plt.show()
```

```python
import pandas as pd

import scipy as sc
from scipy.stats import norm


import matplotlib.pyplot as plt
mu=0
sigma=1
x=sc.arange(mu-4*sigma,mu+4*sigma,0.001)
cdf=norm.cdf(x,loc=mu , scale=sigma)
plt.title("Normal / Gaussian Distribution")
plt.ylim(0,0.6)
plt.ylabel("Normal variable trials")
plt.plot(x,cdf,linewidth=2,color='b')
plt.show()
```
